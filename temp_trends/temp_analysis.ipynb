```python
import numpy as np
```

```python
np.random.seed(0) 

cities= ['Nairobi', 'Mombasa', 'Kisumu', 'Eldoret', 'Nakuru']

mean_temps=[21, 28, 25, 19, 23]

temp_data= np.array([
    np.random.normal(loc=mean, scale=2, size=30).round(1)
    for mean in mean_temps
])
                    
```

```python
temp_data
```

```python
print("Average temperatures per city in degrees celcius:")
for _ in range(5):
    print(f'{cities[_]}- {temp_data.mean(axis=1)[_].round(1)}')
```

```python
print("Min temperatures per city:")
for _ in range(5):
    print(f'{cities[_]}: {temp_data.min(axis=1)[_]}')
```

```python
print("Max temperatures per city:")
for _ in range(5):
    print(f'{cities[_]}: {temp_data.max(axis=1)[_]}')
```

```python
print("Hottest and coldest day overall in degrees celcius.\n")
print(f'The hottest day was at {temp_data.max()} and the coldest at {temp_data.min()}')
```

```python
print("Hottest and coldest day per city.\n")
for _ in range(5):
    print(f'In {cities[_]} the hottest day was at {temp_data.max(axis=1)[_]} and the coldest at {temp_data.min(axis=1)[_]}')
```

```python
std_devs= temp_data.std(axis=1)
indx=np.argmax(std_devs)

print(f'The city which had the most variability is {cities[indx]} with a var of: {std_devs[np.argmax(std_devs)].round(2)}')
```

```python
print("City to city trends.\nDay to day temp differences for NBO and MBS in degree celcius.\n")
print('NBO | MBS| DIFF')
x=0
for x in range(30):
    diff= temp_data[0:2,x][1]-temp_data[0:2,x][0]
    print(f'{temp_data[0:2,x]} {diff.round(1)}')
    x=x+1
    if diff<0:
        print(f'\tNBO was hotter than MBS')
```

```python
print('Days with temperatures below 19 centigrade per city.\n')

for _ in range(5):
    d=temp_data[_]
    d1=d<19
    l=len(d[d1])
    print(f'{cities[_]}: {l} days')
```

```python
print('Days with temperatures above 28 centigrade per city.\n')

for _ in range(5):
    d=temp_data[_]
    d1=d>28
    l=len(d[d1])
    print(f'{cities[_]}: {l} days')
```

```python
labels= np.where(temp_data>28, 'Heat Wave', 
                 np.where(temp_data<19, 'Cold Snap', 'Normal'))
```

```python
def extreme_temperatures_count(labels):
    for i, city in enumerate(cities):
        heat_wave_days= np.sum(labels[i]== "Heat Wave")
        cold_snap_days= np.sum(labels[i]== "Cold Snap")
        print(f"{city}: {heat_wave_days} Heat Wave Days, {cold_snap_days} Cold Snap Days\n")
        print(f'{labels[i]}\n')
        
extreme_temperatures_count(labels)
```

```python
import matplotlib.pyplot as plt

for i, city in enumerate (cities):
    plt.plot(temp_data[i], label= city)
    
    plt.title("30-Day Temperature Trends")
    plt.xlabel("Day")
    plt.ylabel("Temperature (Centigrade)")
    plt.legend()
    plt.show()
```
